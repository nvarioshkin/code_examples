```python
data = pd.read_excel('automation per thous workers.xlsx', usecols = 'A:AB')
data.drop(['market'], axis = 1, inplace = True)
data.dropna(axis = 0, how = 'any', subset = ['country', 'industry'], inplace = True)
```

```python
data
```

```python
short = ['     A-B-Agriculture, forestry, fishing', '     C-Mining and quarrying', '          10-12-Food and beverages', 
        '          13-15-Textiles', '          16-Wood and furniture', '          17-18-Paper', '          19-22-Plastic and chemical products', 
        '          23-Glass, ceramics, stone, mineral products (non-auto', '          24-28-Metal', '          26-27-Electrical/electronics', '          29-Automotive', '          30-Other vehicles', '          91-All other manufacturing branches', '     E-Electricity, gas, water supply', '     F-Construction', '     P-Education/research/development', '     90-All other non-manufacturing branches', '     99-Unspecified']
```

```python
short
```

```python
df = data.set_index(keys = ['country', 'industry']).stack().reset_index()
df.columns = ['country', 'industry', 'year', 'robots']
df = df[df['industry'].isin(short)]
df['industry'] = df['industry'].str.strip(' ')
```

```python
df
```

```python
df_usa = df[df['country'] == 'USA'].reset_index(drop = True)
df_other = df[df['country'] != 'USA'].reset_index(drop = True)

df_europe = df_other.groupby(by = ['industry', 'year'], as_index = False).sum()
df_europe.insert(0, 'country', 'Europe')

df_short = pd.concat([df_usa, (df_europe)])
df_short.reset_index(drop = True, inplace = True)

df_short['industry'] = df_short['industry'].str.rpartition(sep = '-')[2]

df_short.loc[df_short['industry'] == 'auto', 'industry'] = 'Glass, ceramics, stone, mineral products'
df_short.loc[df_short['industry'] == 'manufacturing branches', 'industry'] = 'All other non-manufacturing branches'
```

```python
df_short
```

```python
labels = list(df_short['industry'].unique())
labels.sort()
numbers = [1, 16, 17] + list(range(2, 15)) + [18] + [15]
map_dict = dict(zip(labels, numbers))
map_dict_rev = dict(zip(numbers, labels))


df_short['ind'] = df_short['industry'].map(map_dict)

label_dict = {'ind' : map_dict_rev}
```

```python
df_short
```

```python
df_short.to_stata('short.dta', write_index = False, value_labels = label_dict)
```

```python
foo_main = data_1['soc1'].drop_duplicates().str.split('_', expand = True)
foo_main = foo_main.reset_index(drop = True)
foo_main['key'] = foo_main.index
foo_main = foo_main.melt(id_vars = 'key').dropna().drop(['variable'], axis = 1)
foo_main = foo_main.sort_values('value').reset_index(drop = True)
foo_main = foo_main.rename({'value' : 'OCC_CODE'}, axis = 1)
foo_main = foo_main[['OCC_CODE', 'key']] 


crosswalk = data[['OCC_CODE', 'OCC_TITLE']].drop_duplicates()
crosswalk = crosswalk.merge(foo_main, on = 'OCC_CODE',  how = 'left')
crosswalk['cross_key'] = crosswalk.index
```

```python
def changer(string) : 
    
    length = len(string)
    
    dum = crosswalk[['OCC_CODE', 'cross_key']].copy()
    dum['OCC_CODE'] = dum['OCC_CODE'].map(lambda x : x[:-length] + string)
    dum = dum.merge(foo_main, on = 'OCC_CODE', how = 'left').drop(['OCC_CODE'], axis = 1)
    crosswalk[['cross_key', 'key']] = crosswalk[['cross_key', 'key']].fillna(dum)
    
    
```

```python
l = ['0', '00', '000', '0000', 'X', 'XX', 'XXX', 'XXXX', 'Y', 'YY', 'YYY', 'YYYY']

for item in l : 
    changer(item)
```

```python
foo_main2 = data_1[['soc1']].drop_duplicates().reset_index(drop = True)
foo_main2 = foo_main2.reset_index(drop = True)
foo_main2['key'] = foo_main2.index

crosswalk = crosswalk.merge(foo_main2, on = 'key', how = 'left')
crosswalk = crosswalk[['OCC_CODE', 'soc1']]
crosswalk.to_stata('crosswalk.dta', write_index = False)
```
